# 2018 인구이동

#### 전처리 1. 파일 불러오기

```python
seoul_pop_2018 = pd.read_csv("C:\\hojun13\\Pythonwork\\Analysis\\Data\\data\\부동산 파일\\프로젝트\\시군구별_이동자수_20230414083900.csv", encoding='cp949')
seoul_pop_2018
```

#### 전처리 2. 알아보기 좋게 표 정리하기

```python
# 2018로 컬럼 이름 바꾸기
for a in range(1, 4):
    seoul_pop_2018.rename(columns={seoul_pop_2018.columns[a] : '2018'}, inplace=True)
```

```python
# 데이터를 알아보기 편하게 이름 바꾸기
seoul_pop_2018.columns.values[1]='2018 총전입'
seoul_pop_2018.columns.values[2]='2018 총전출'
seoul_pop_2018.columns.values[3]='2018 순이동'
seoul_pop_2018.columns.values[0]='자치구별'
```

```python
#필요없는 행 삭제하기
seoul_pop_2018.drop(seoul_pop_2018.index[0],inplace=True)
```

```python
#계산을 위해 데이터 타입 바꾸기
seoul_pop_2018['2018 총전입']=pd.to_numeric(seoul_pop_2018['2018 총전입'])
seoul_pop_2018['2018 총전출']=pd.to_numeric(seoul_pop_2018['2018 총전출'])
seoul_pop_2018['2018 순이동']=pd.to_numeric(seoul_pop_2018['2018 순이동'])
```

```python
# 인구이동률 컬럼 추가하기
seoul_pop_2018['2018 인구이동률(%)']=round(seoul_pop_2018['2018 순이동']/(seoul_pop_2018['2018 총전입']+seoul_pop_2018['2018 총전출']),3)*100
```

```python
# 자료 저장하기
seoul_pop_2018.to_csv("C:\\hojun13\\Pythonwork\\Analysis\\Data\\data\\부동산 파일\\프로젝트\\2018인구이동.csv",encoding='utf-8')
```
